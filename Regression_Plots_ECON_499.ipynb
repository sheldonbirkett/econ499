```python
min_dist_km = acs_2005_2019.iloc[:,78:92]
min_dist_km_mean = min_dist_km.mean()
```

```python
min_dist_quant = min_dist_km.quantile([0.25,0.75])
```

```python
cols = list(min_dist_quant.columns.values)
for i in range(len(cols)):
    min_dist_quant[cols[i]] = min_dist_km_mean[i] - min_dist_quant[cols[i]] 
```

```python
min_dist_quant = min_dist_quant.abs()
```

```python
min_dist_quant
```

```python
min_dist_quant = min_dist_quant.to_numpy()
```

```python
acs_medval_trends = acs_2005_2019[["YEAR", "medval","lowval","upval", "min_dist_2015_sim"]].copy()
```

```python
acs_medval_trends["Fire_4km"] = (acs_medval_trends["min_dist_2015_sim"]<=4000)
```

```python
acs_medval_trends["lnmedval"] = np.log(acs_medval_trends["medval"])
acs_medval_trends["lnlowval"] = np.log(acs_medval_trends["lowval"])
acs_medval_trends["lnupval"] = np.log(acs_medval_trends["upval"])
```

```python
acs_medval_trends_group = acs_medval_trends.groupby(["YEAR","Fire_4km"]).mean()
```

```python
acs_medval_trends_group = acs_medval_trends_group.unstack(level = 1)
```

```python
acs_medval_trends_group["medval", True]
```

```python
acs_medval_trends_group = acs_medval_trends_group.reset_index()
```

```python
acs_medval_trends_group["YEAR"]
```

```python
plt.rcParams['figure.figsize'] = (20,10)
plt.style.use('fivethirtyeight')

fig, ax = plt.subplots()

ax.plot(acs_medval_trends_group["YEAR"],acs_medval_trends_group["medval", False],label = r"Not Within 4km from a 2015 Wildfire Perimeter", color='teal')
ax.plot(acs_medval_trends_group["YEAR"],acs_medval_trends_group["medval", True],label = r"Within 4km from a 2015 Wildfire Perimeter", color='purple')
# ax.plot(df["Cutoff_Value"], df["beta"], label = r"Coefficient $\beta_1$ 4 km Cutoff", color='teal', alpha=.6, linestyle = 'dashed')
# ax.fill_between(df["Cutoff_Value"], df["Lower_Bound_CI"], df["Upper_Bound_CI"], color='purple', alpha=.2, label = '95% Confidence Interval')
# ax.fill_between(df["Cutoff_Value"], df["lower_CI"], df["upper_CI"], color='teal', alpha=.2, label = '95% Confidence Interval')
ax.set_title('Pretrends Median Property Value by Within 4km from a 2015 Wildfire Perimeter')
ax.set_ylabel(r'Average Median Property Value')
ax.set_xlabel(r'Periods')

ax.legend(loc = "upper right")

#fig.savefig('medval_trends.png', bbox_inches='tight')
```
